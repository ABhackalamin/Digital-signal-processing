```python
import numpy as np
from scipy.signal import iirfilter, lfilter
import matplotlib.pyplot as plt

# Define filter parameters
order = 4    # Filter order
cutoff_freq = 0.2  # Cutoff frequency in normalized units (0.0 to 1.0)
sampling_freq = 100  # Sampling frequency in Hz

# Design the filter
b, a = iirfilter(order, cutoff_freq, btype='low', analog=False, fs=sampling_freq)

# Generate a test signal (for example, a sine wave)
t = np.linspace(0, 1, sampling_freq, endpoint=False)
input_signal = np.sin(2 * np.pi * 5 * t)  # 5 Hz sine wave

# Apply the filter to the signal
output_signal = lfilter(b, a, input_signal)

# Plot the original and filtered signals
plt.figure(figsize=(10, 4))
plt.subplot(2, 1, 1)
plt.plot(t, input_signal)
plt.title('Original Signal')
plt.subplot(2, 1, 2)
plt.plot(t, output_signal)
plt.title('Filtered Signal')
plt.tight_layout()
plt.show()
```
